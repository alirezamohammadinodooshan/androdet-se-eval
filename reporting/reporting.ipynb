```python
import pandas as pd
```

# Results for the family study 

### Accuracy results of evaluating each of the families

```python
families_study_dir = '../families-study/'
families_study_atm_logs = pd.read_csv(families_study_dir + 'atm_results.csv', index_col='family')
families_study_moa_logs = pd.read_csv(families_study_dir + 'moa_results.csv', index_col='family')

families = [] 
families_study_atm_accuracy_results = []
families_study_moa_accuracy_results = []

for family in families_study_atm_logs.index.values:
    family_atm_tp = families_study_atm_logs.loc[family]['tp']
    family_atm_tn = families_study_atm_logs.loc[family]['tn']
    family_atm_fp = families_study_atm_logs.loc[family]['fp']
    family_atm_fn = families_study_atm_logs.loc[family]['fn']

    family_moa_tp = families_study_moa_logs.loc[family]['tp']
    family_moa_tn = families_study_moa_logs.loc[family]['tn']
    family_moa_fp = families_study_moa_logs.loc[family]['fp']
    family_moa_fn = families_study_moa_logs.loc[family]['fn']

    family_atm_accuracy = (family_atm_tp + family_atm_tn)/\
    (family_atm_tp + family_atm_tn + family_atm_fp + family_atm_fn)

    family_moa_accuracy = (family_moa_tp + family_moa_tn)/\
    (family_moa_tp + family_moa_tn + family_moa_fp + family_moa_fn)
    
    families.append(family)
    families_study_atm_accuracy_results.append(family_atm_accuracy)    
    families_study_moa_accuracy_results.append(family_moa_accuracy)
    
    
print("families,atm_study_accuracy,moa_study_accuracy")
for i in range(len(families)):
    print("%s,%.2f,%.2f" % (families[i], families_study_atm_accuracy_results[i], families_study_moa_accuracy_results[i]))
```

### Average results for the families study

```python
atm_total_values = families_study_atm_logs.sum()
atm_total_tp = atm_total_values['tp']
atm_total_tn = atm_total_values['tn']
atm_total_fp = atm_total_values['fp']
atm_total_fn = atm_total_values['fn']
atm_accuracy = (atm_total_tp + atm_total_tn)/\
(atm_total_tp + atm_total_tn + atm_total_fp + atm_total_fn)
atm_f1_score = (2 * atm_total_tp)/\
(2 * atm_total_tp + atm_total_fp + atm_total_fn)

moa_total_values = families_study_moa_logs.sum()
moa_total_tp = moa_total_values['tp']
moa_total_tn = moa_total_values['tn']
moa_total_fp = moa_total_values['fp']
moa_total_fn = moa_total_values['fn']
moa_accuracy = (moa_total_tp + moa_total_tn)/\
(moa_total_tp + moa_total_tn + moa_total_fp + moa_total_fn)
moa_f1_score = (2 * moa_total_tp)/\
(2 * moa_total_tp + moa_total_fp + moa_total_fn)

print("Average accuracy for ATM study : {}".format(atm_accuracy))
print("Average accuracy for MOA study : {}".format(moa_accuracy))
print("Average fscore for ATM study : {}".format(atm_f1_score))
print("Average fscore for MOA study : {}".format(moa_f1_score))  
```

# Results for the random study 

### Accuracy results for each of the 100 folds

```python
random_study_dir = '../random-study/'
random_study_atm_logs = pd.read_csv(random_study_dir + 'atm_results.csv', index_col='cross_validation')####
random_study_moa_logs = pd.read_csv(random_study_dir + 'moa_results.csv', index_col='cross_validation')####

cross_validations = [] 
random_study_atm_accuracy_results = []
random_study_moa_accuracy_results = []

for cross_validation in random_study_atm_logs.index.values:
    cross_validation_atm_tp = random_study_atm_logs.loc[cross_validation]['tp']
    cross_validation_atm_tn = random_study_atm_logs.loc[cross_validation]['tn']
    cross_validation_atm_fp = random_study_atm_logs.loc[cross_validation]['fp']
    cross_validation_atm_fn = random_study_atm_logs.loc[cross_validation]['fn']

    cross_validation_moa_tp = random_study_moa_logs.loc[cross_validation]['tp']
    cross_validation_moa_tn = random_study_moa_logs.loc[cross_validation]['tn']
    cross_validation_moa_fp = random_study_moa_logs.loc[cross_validation]['fp']
    cross_validation_moa_fn = random_study_moa_logs.loc[cross_validation]['fn']

    cross_validation_atm_accuracy = (cross_validation_atm_tp + cross_validation_atm_tn)/\
    (cross_validation_atm_tp + cross_validation_atm_tn + cross_validation_atm_fp + cross_validation_atm_fn)

    cross_validation_moa_accuracy = (cross_validation_moa_tp + cross_validation_moa_tn)/\
    (cross_validation_moa_tp + cross_validation_moa_tn + cross_validation_moa_fp + cross_validation_moa_fn)
    
    cross_validations.append(cross_validation)
    random_study_atm_accuracy_results.append(cross_validation_atm_accuracy)    
    random_study_moa_accuracy_results.append(cross_validation_moa_accuracy)
    
    
print("cross_validation_no,atm_study_accuracy,moa_study_accuracy")
for i in range(len(cross_validations)):
    print("%s,%.2f,%.2f" % (cross_validations[i], random_study_atm_accuracy_results[i], random_study_moa_accuracy_results[i]))
```
